```python
import matplotlib.pyplot as plt
import numpy as np

# Předpokládaná data
velikosti_matice = np.array([2, 4, 8, 16, 32])
prumerny_cas = np.array([0.5, 2.0, 5.0, 10.0, 20.0])

# Vytvoření grafu
plt.figure(figsize=(10, 6))
plt.plot(velikosti_matice, prumerny_cas, marker='o')

# Přidání popisků
plt.title('Průměrný čas v závislosti na velikosti čtvercové matice')
plt.xlabel('Velikost čtvercové matice')
plt.ylabel('Průměrný čas (s)')

# Zobrazit mřížku
plt.grid(True)

# Zobrazit graf
plt.show()
```
