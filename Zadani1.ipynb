 Spočítání integrálu
 

```python
#řešení určitého integrálu pomocí scipy

import scipy
from scipy import  integrate
import time


t1 = time.time()

def f(x):
  return 3 * (x**2) +1
result = integrate.quad(f, -3, 2)

t2= time.time() - t1

print(result)
print(t2)



#řešení určitého integrálu pomocí pythonu
# 3 * (x **2) +1

t3 = time.time()
def f(a, x, n, c ):
  if c == 0:
    integral = a * (1/(n+1)) * (x ** (n+1))
  else:
    integral = a * (1/(n+1)) * (x ** (n+1)) + (c * x)
  return integral

b_integral = f(3, 2, 2, 1)

a_integral = f(3, -3, 2, 1)

vysledek = b_integral - a_integral
t4= time.time() -t3

print(vysledek)
print(t4)


# srovnání rychlosti

print(f'python je {t4}')
print(f'scipy je {t2}')
print(f'Scipy je zde {t4/t2} rychlejší ')
```

```python
# deterninant matici 

import time 
import numpy as np 

start = time.time()
matice = np.array([
    [3, -5, 2],
    [7, 2, -3],
    [3, -2, 1]])

vysledek = np.linalg.det(matice)

end = time.time() - start

print(f'Determinant je {vysledek}, čas je {end}') 
```
